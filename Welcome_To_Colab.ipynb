MERGE IN DATAFRAME



```
#import libraries
import pandas as pd

#create DF's
df1=pd.DataFrame({"A":[1,2,4],"B":[11,21,31]})
df2=pd.DataFrame({"A":[1,2,4],"B":[11,21,31]})
#perform merge
pd.merge(df2,df2,on="A")
```

```
#USING HOW

```

```
#inner
df1=pd.DataFrame({"A":[1,2,4,5],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,4,5],"C":[10,20,30,40]})
#perform merge
pd.merge(df1,df2,on="A",how="inner")
```

```
#outer
df1=pd.DataFrame({"A":[1,2,4,5],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,4,5],"C":[10,21,30,40]})
#perform merge
pd.merge(df1,df2,on="A",how="outer")
```

```
#left
df1=pd.DataFrame({"A":[1,2,4,5],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,4,6],"C":[10,21,30,40]})
#perform merge
pd.merge(df1,df2,on="A",how="left")
```

```
#indexing
```

```
df1=pd.DataFrame({"A":[1,2,4,4],"B":[11,21,31,41]})
df2=pd.DataFrame({"A":[1,2,4,6],"B":[10,21,30,40]})

pd.merge(df1,df2,left_index=True,right_index=True,suffixes=("_id","_rollno"))
```
